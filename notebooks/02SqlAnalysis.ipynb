```python
import pandas as pd
from sqlalchemy import create_engine
import os

cleaned_path = r"C:\Projects\Olist\data\cleaned"

# Update these with your actual Postgres credentials
engine = create_engine("postgresql://postgres:LESSpostgres@localhost:5432/olistDB")

tables = {
    "orders": "orders_clean.csv",
    "order_items": "order_items_clean.csv",
    "order_payments": "payments_clean.csv",
    "order_reviews": "reviews_clean.csv",
    "customers": "customers_clean.csv",
    "products": "products_clean.csv",
    "sellers": "sellers_clean.csv",
    "geolocation": "geolocation_clean.csv"
}

for table_name, file in tables.items():
    df = pd.read_csv(os.path.join(cleaned_path, file))
    df.to_sql(table_name, engine, if_exists="replace", index=False, schema="public")
    print(f"Loaded {table_name}: {len(df)} rows")
```
